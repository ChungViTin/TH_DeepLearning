## Phân tích và Xử lý dữ liệu Tuyển sinh Đại học - Lab 1

1. **Phân loại dữ liệu:**
   - **Định tính (Qualitative):** GT (Nominal), DT (Nominal), KV (Ordinal), KT (Nominal), XL1, XL2, XL3 (Ordinal), KQXT (Nominal).
   - **Định lượng (Quantitative):** T1-T6, L1-L6, H1-H6, S1-S6, V1-V6, X1-X6, D1-D6, N1-N6, DH1-DH3 (Ratio scale).

2. **Thang đo:**
   - Điểm số: Thang đo tỷ lệ (Ratio).
   - Khu vực (KV): Thang đo thứ bậc (Ordinal).
   - Khối thi (KT), Dân tộc (DT), Giới tính (GT): Thang đo định danh (Nominal).

```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/dulieuxettuyendaihoc.csv'

try:
    df = pd.read_csv(url)
    print("Tải dữ liệu thành công từ URL.")

    print("10 dòng đầu tiên:")
    display(df.head(10))
    print("\n10 dòng cuối cùng:")
    display(df.tail(10))
except Exception as e:
    print(f"Không thể tải dữ liệu từ URL. Lỗi: {e}")
```

```python
# 4. Thống kê và xử lý dữ liệu thiếu cho cột 'DT' (Dân tộc)
print("Tần số dữ liệu 'DT' trước khi xử lý:")
print(df['DT'].value_counts(dropna=False))

# Điền giá trị 0 cho dữ liệu thiếu
df['DT'] = df['DT'].fillna(0)

print("\nTần số dữ liệu 'DT' sau khi xử lý:")
print(df['DT'].value_counts())
```

```python
# 5. Thống kê và xử lý dữ liệu thiếu cho 'T1' bằng Mean
print(f"Số lượng giá trị thiếu trong T1: {df['T1'].isnull().sum()}")

t1_mean = df['T1'].mean()
df['T1'] = df['T1'].fillna(t1_mean)

# 6. Xử lý dữ liệu thiếu cho các biến điểm số còn lại
score_cols = ['T1','L1','H1','S1','V1','X1','D1','N1',
              'T2','L2','H2','S2','V2','X2','D2','N2',
              'T6','L6','H6','S6','V6','X6','D6','N6',
              'DH1','DH2','DH3']

for col in score_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mean())

print("Đã xử lý xong dữ liệu thiếu cho tất cả các cột điểm.")
```

```python
# 7. Tạo biến TBM1, TBM2, TBM3
# Công thức: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
# (Giả định T và V hệ số 2 theo đề bài gợi ý T2 + V2...)
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10

display(df[['TBM1', 'TBM2', 'TBM3']].head())
```

```python
# 8. Phân loại XL1, XL2, XL3
def phan_loai(diem):
    if diem < 5.0: return 'Y'
    elif diem < 6.5: return 'TB'
    elif diem < 8.0: return 'K'
    elif diem < 9.0: return 'G'
    else: return 'XS'

for i in [1, 2, 3]:
    df[f'XL{i}'] = df[f'TBM{i}'].apply(phan_loai)

display(df[['XL1', 'XL2', 'XL3']].head())
```

```python
# 9. Chuyển sang thang điểm 4 (Min-Max Normalization)
# Công thức: (x - min) / (max - min) * 4
for i in [1, 2, 3]:
    col = f'TBM{i}'
    df[f'US_TBM{i}'] = ((df[col] - 0) / (10 - 0)) * 4

display(df[['US_TBM1', 'US_TBM2', 'US_TBM3']].head())
```

```python
# 10. Tạo biến KQXT
def xet_tuyen(row):
    kt = row['KT']
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']

    if kt in ['A', 'A1']:
        diem = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B':
        diem = (dh1 + dh2*2 + dh3) / 4
    else:
        diem = (dh1 + dh2 + dh3) / 3

    return 1 if diem >= 5.0 else 0

df['KQXT'] = df.apply(xet_tuyen, axis=1)

# 11. Lưu trữ dữ liệu
df.to_csv('/content/drive/MyDrive/TH_DeepLearning/processed_dulieuxettuyendaihoc.csv', index=False)
print("Đã lưu file: processed_dulieuxettuyendaihoc.csv")
```

# BÀI THỰC HÀNH: XỬ LÝ DỮ LIỆU

### PHẦN 1: THIẾU THÔNG TIN?
1. Chuyển đổi ký tự lạ về NaN.
2. Thống kê tỷ lệ dữ liệu thiếu.
3. Trực quan hóa bằng Heatmap.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tải dữ liệu
url = 'https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/heart_disease_dataset.csv'
df = pd.read_csv(url)

# Chuyển '?' và khoảng trắng thành NaN
df.replace(['?', ' '], np.nan, inplace=True)

# Thống kê tỷ lệ thiếu
missing_percent = df.isnull().mean() * 100
print("Tỷ lệ dữ liệu thiếu trên từng cột (%):")
print(missing_percent[missing_percent > 0])

# Vẽ Heatmap trực quan hóa dữ liệu thiếu
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title('Heatmap: Vị trí dữ liệu bị khuyết (Màu vàng là NaN)')
plt.show()
```

### PHẦN 2: LOẠI BỎ VÀ ĐIỀN GIÁ TRỊ
1. Xóa dòng thiếu >= 3 thuộc tính.
2. Điền Median cho biến liên tục, Mode cho biến phân loại.

```python
# Loại bỏ dòng thiếu từ 3 thuộc tính trở lên
limit = df.shape[1] - 3
df_lab = df.dropna(thresh=limit).copy()

df_lab = df_lab.dropna(subset=['state'])

# Điền giá trị (Imputation)
# Biến liên tục
num_vars = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
for col in num_vars:
    df_lab[col] = pd.to_numeric(df_lab[col], errors='coerce')
    df_lab[col] = df_lab[col].fillna(df_lab[col].median())

# Biến phân loại
cat_vars = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
for col in cat_vars:
    mode_val = df_lab[col].mode()[0]
    df_lab[col] = df_lab[col].fillna(mode_val)

print(f"Số lượng missing values sau xử lý: {df_lab.isnull().sum().sum()}")
plt.figure(figsize=(10, 6))
sns.heatmap(df_lab.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title('Heatmap: Đã làm sạch 100% Missing Values')
plt.show()
```

### PHẦN 3: OUTLIER? REMOVE?
Sử dụng phương pháp IQR để loại bỏ các giá trị dị biệt cho 3 biến: trestbps, chol, thalach.

```python
target_cols = ['trestbps', 'chol', 'thalach']

# Vẽ Box-plot trước khi xử lý
plt.figure(figsize=(12, 4))
for i, col in enumerate(target_cols):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=df_lab[col])
    plt.title(f'Trước: {col}')
plt.tight_layout()
plt.show()

# Hàm loại bỏ Outlier bằng IQR
def remove_outliers_iqr(data, columns):
    df_final = data.copy()
    for col in columns:
        Q1 = df_final[col].quantile(0.25)
        Q3 = df_final[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        # Lọc dữ liệu
        df_final = df_final[(df_final[col] >= lower) & (df_final[col] <= upper)]
    return df_final

df_lab = remove_outliers_iqr(df_lab, target_cols)

# Vẽ lại Box-plot sau khi xử lý
plt.figure(figsize=(12, 4))
for i, col in enumerate(target_cols):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=df_lab[col], color='lightgreen')
    plt.title(f'Sau: {col}')
plt.tight_layout()
plt.show()
```

### PHẦN 4: CÂN BẰNG DỮ LIỆU
1. Nhị phân hóa biến `state` thành `target`.
2. Sử dụng SMOTE để cân bằng lớp.
3. Lưu file `ready_for_DL.csv`.

```python
from imblearn.over_sampling import SMOTE

# Biến đổi state thành target (0: No, 1: Yes)
df_lab['target'] = df_lab['state'].apply(lambda x: 1 if float(x) > 0 else 0)

# Vẽ Bar-chart kiểm tra mất cân bằng
plt.figure(figsize=(6,4))
sns.countplot(x='target', data=df_lab)
plt.title('Tỉ lệ lớp Target (0 vs 1)')
plt.show()

# SMOTE yêu cầu dữ liệu dạng số, ta thực hiện Encoding nhanh
df_encoded = pd.get_dummies(df_lab.drop(columns=['state']), drop_first=True)
X = df_encoded.drop(columns=['target'])
y = df_encoded['target']

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Trực quan hóa sau cân bằng
plt.figure(figsize=(6,4))
sns.countplot(x=y_res)
plt.title('Phân phối lớp sau khi dùng SMOTE')
plt.show()

# Lưu kết quả
ready_df = pd.concat([X_res, y_res], axis=1)
ready_df.to_csv('heart_disease.csv', index=False)
print("Đã lưu bộ dữ liệu chuẩn hóa thành công: ready_for_DL.csv")
```
